```python

import pandas as pd
import numpy as np
d1 = {'First value':[10, 90, np.nan, np.nan],
		'Second value': [np.nan, 4, 6, np.nan],
		'Third Score':[40, 80, np.nan, 97]}
d = pd.DataFrame(d1)
d.isnull()
```

```python
d.fillna(0)
```

```python
d.fillna(method ='pad')
```

```python
d.fillna(method ='bfill')
```
